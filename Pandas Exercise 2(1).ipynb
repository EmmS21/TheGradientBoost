## Question 3

What is the average size of houses that are between 50k and 100k Naira. Size in this context refers to the number of bedrooms and baths in each house

```python
df.baths.unique()
```

```python
#answer here

# Since the baths and bedroom columns are not numeric, it it necessary to clean and convert the to numeric
# in order to complete thhe task

df['bedrooms'] = df['bedrooms'].apply(lambda x: x.replace('bed', '').replace('br', '').replace(' ', '0') if pd.notna(x) else x).astype(np.float64)
df['baths'] = df['baths'].apply(lambda x: x.replace('bath', '').replace('br', '').replace(' ', '0') if pd.notna(x) else x).astype(np.float64)

df.baths.unique()
```

```python
df.loc[df['price'].between(50000, 100000)][['bedrooms', 'baths']].mean()
```
