- Net bilateral aid flows from DAC donors, United States or

Save this as western_dac_donors

# Question 2

Great, now select the rows where the indicator_name relates to Net officual flows from UN agencies, save this subset as un_agencies_donor

```python
#answer here

un_agencies_donor = df[df['indicator_name'].apply(lambda x: 'Net official flows from UN agencies,' in x)]
un_agencies_donor.head()
```

# Question 3

Select all the rows in un_agencies_donor where the year is greater than or equal to 2015, sort these rows by <strong>value</strong> in descending order ie. starting with the largest value

```python
#answer here

un_agencies_donor = un_agencies_donor[un_agencies_donor['year'].apply(lambda x: x>2015 or x==2015)].sort_values(by='value', ascending=False)
un_agencies_donor.head()
```

# Question 4
How much money did Nigeria receive from 
- i.) WHO in 2016 and 2017
- ii.) bilateral aid flows from DAC donor, United Kingdom in 2016 and 2017

```python
import numpy as np

WHO = df[df['indicator_name'] == 'Net official flows from UN agencies, WHO (current US$)']#
WHO = WHO[WHO['year'].apply(lambda x: x==2016 or x==2017)]
WHO_funds = WHO['value'].agg(np.sum)
print('The total mony received by Nigeria from the WHO from 2016 to 2017 is {}'.format(WHO_funds))
```

```python
uk = df[df['indicator_name'] == 'Net bilateral aid flows from DAC donors, United Kingdom (current US$)']
uk = uk[uk['year'].apply(lambda x: x==2016 or x==2017)]
uk_funds = uk['value'].agg(np.sum)
print('The total mony received by Nigeria from bilateral aid flows from DAC donor, United Kingdom in 2016 and 2017 is {}'.format(uk_funds))
```
